```
# importing libraries 
import pandas as pd
import seaborn as ss
import matplotlib.pyplot as plt
```

##Ques 1

```
# importing datasets of weather forecast
df=pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv")

# printing first 5 rows of dataset
df.head()f
```

```
# group by winddir9am
dfg=df.groupby(by="WindDir9am").agg(["mean","sum"])

   
```

```
# Distribution plot of Humidity3pm
ss.distplot(df["Humidity3pm"])
ss.despine()
```

```
# Distribution plot of Maxtemp
ss.distplot(df["MaxTemp"])
plt.title("Distribution curve for Maxtemp")
ss.despine()
```

```
# Countplot of Raintoday
ss.countplot(x="RainToday",data=df)
plt.title("Countplot of raintoday").title
ss.despine()
```

```
# Count plot of Rain tomorrow
ss.countplot(x="RainTomorrow",data=df)
plt.title("countplot of Raintomorrow")
ss.despine()
```

```
dfg["Rainfall"]["mean"].plot(kind="bar")
```

```
dfg["Rainfall"]["sum"].plot(kind="barh")
```

```
dfg["Evaporation"]["mean"].plot(kind="pie")
```

```
dfg["Sunshine"]["sum"].plot(kind="pie")
```

```
df["Temp9am"].plot(kind="hist")
```

```
# Boxplot for Raintoday Vs Mintemp
ss.boxplot(x="RainToday",y="MinTemp",data=df)
plt.title("Boxplot for Rain today vS MinTemp")
ss.despine()
```

```
# boxplot of winddir9am vs maxtemp
ss.boxplot(x="WindDir9am",y="MaxTemp",data=df)
plt.title("Boxplot for Winddir9am VS Maxtemp")
ss.despine()
```

```
# Violinplot for raintoday vs mintemp
ss.violinplot(x="RainToday",y="MinTemp",data=df)
plt.title("Violinplot for Rain today vS MinTemp")
ss.despine
```

```
# scatterplot for cloud9am Vs humidity3pm
ss.jointplot(x="Cloud9am",y="Humidity3pm",data=df,alpha=0.2)
ss.despine()
```
